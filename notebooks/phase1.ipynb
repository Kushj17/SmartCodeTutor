```python
import json
import time
import random
from datasets import load_dataset
import google.genai as genai
from tqdm import tqdm
```

```python

client = genai.Client(api_key="YOUR_API_KEY")
MODEL = "gemini-3.1-flash-lite-preview"


MAX_SAMPLES_PER_LANG = 5000
OUTPUT_FILE = "../data/dataset.json"
TEMP_FILE = "../data/dataset_temp.json"

LANGUAGES = ["python", "java", "javascript", "go"]
```

```python

def generate_explanation(code):
    prompt = f"""
Explain the following code in a beginner-friendly way.

Rules:
- Use simple language
- Break into steps
- Avoid technical jargon
- keep it under 50 words

Code:
{code}
"""

    for attempt in range(3):
        try:
            response = client.models.generate_content(
                model=MODEL,
                contents=prompt
            )
            return response.text.strip()

        except Exception as e:
            print(f"Retry {attempt+1}:", e)
            time.sleep(2 + attempt)

    return None

```

```python

def is_valid_code(code):
    if not code:
        return False
    if len(code) < 20:
        return False
    if len(code) > 500:
        return False
    if code.count("\n") < 1:
        return False
    return True


def save_progress(data):
    with open(TEMP_FILE, "w") as f:
        json.dump(data, f, indent=4)

```

```python

data = []

for lang in LANGUAGES:
    print(f"\nProcessing language: {lang}")

    dataset = load_dataset("code_search_net", lang, split="train")

    count = 0

    for item in tqdm(dataset):
        if count >= MAX_SAMPLES_PER_LANG:
            break

        code = item.get("func_code_string")
        doc = item.get("func_documentation_string")


        if not code:
            continue

        if doc and len(doc.strip()) > 20:
            explanation = doc
        else:
            explanation = generate_explanation(code)

        if explanation:
            data.append({
                "input": code,
                "output": explanation,
                "language": lang
            })

            count += 1

            if len(data) % 10 == 0:
                save_progress(data)

        time.sleep(random.uniform(1, 2))


with open(OUTPUT_FILE, "w") as f:
    json.dump(data, f, indent=4)

print(f"\nDone! Total samples: {len(data)}")
```
